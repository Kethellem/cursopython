COLUNA GENERO

Nessa coluna temos apenas 0.48% dos dados faltantes pouco para uma base tão grande

```python
df[df['Genero'].isnull()]
```

Excluindo valores faltantes da coluna Genero

```python
# Excluindo
df.dropna(subset=['Genero'], inplace=True)
#InplaceTrue faz com que as alterações sejam feitas diretamente no DataFrame original
```

```python
df['Genero'].isnull().sum()
```

a exclusão de dados faltantes na coluna Genero nao gera impacto na nossa base de dados já que a informação principal que queremos e Churn

COLUNA CustomerID

```python
#Removendo coluna customerID
df.drop(columns=['customerID'], inplace=True)
```

Excluindo coluna customerID pois não são dados relevantes dentro do nosso modelo

COLUNA PHONESERVICE


preenchendo os nulos da coluna phoneservice com a moda (valor mais frequente)

```python
# Encontrar o valor mais frequente (moda)
moda_PhoneService = df['PhoneService'].mode()

# Preencher os nulos com a moda
df['PhoneService'].fillna(moda_PhoneService, inplace=True)
```

```python
# Verificando se ainda resta dandos faltantes
(df.isnull().sum() / len(df)) * 100
```

“Realizei o tratamento dos dados faltantes.
Na coluna PhoneService, os valores nulos foram preenchidos com o valor mais frequente encontrado no dataset.

COLUNA PAGAMENTO_MENSAL


Fazendo o tratamento dos dados da coluna pagamento_mensal com Média e Mediana para preencher o execesso de registros

```python
# Analisando Média
df['Pagamento_Mensal'].mean()
```

```python
# Analisando Mediana
df['Pagamento_Mensal'].median()
```

SUBISTITUINDO PELA MEDIANA

```python
mediana_Pagamento_Mensal = df['Pagamento_Mensal'].median()
# Substituindo os valores faltantes na coluna de Pagamento Mensal
df['Pagamento_Mensal'].fillna(mediana_Pagamento_Mensal, inplace=True)
```

```python
df['Pagamento_Mensal'].isnull().sum()
```

```python
# Verificando colunas se ainda possui dados faltantes
(df.isnull().sum() / len(df)) * 100
```

Para a coluna Pagamento_Mensal, utilizei a mediana como forma de preenchimento, considerando que essa estratégia reduz impactos de possíveis outliers.


CORRIGINDO DADOS INCONSISTENTES

```python
df['Servico_Internet'].unique()
```

```python
#Transformando todos os dados da coluna Servico_Internet em caixa alta
df['Servico_Internet'] = df['Servico_Internet'].str.upper()
```

```python
#Utilizando a função replace
df['Servico_Internet'] = df['Servico_Internet'].replace('DSL', 'DLS')
```

```python
#Visualizando alterações
df['Servico_Internet'].unique()
```

Os dados errados foram subistituido pela correção em letras maiusculas utilizando o comando upper

PADRONIZANDO COLUNAS TODAS EM PORTUGUÊS

```python
#Utilizando a função rename
df.rename(columns={'Dependents': 'Dependentes'}, inplace=True)
df.rename(columns={'PhoneService': 'Servico_Telefonico'}, inplace=True)
df.rename(columns={'StreamingTV': 'Transmissao_TV'}, inplace=True)
df.rename(columns={'PaymentMethod': 'Forma_Pagamento'}, inplace=True)
```
