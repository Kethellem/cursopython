Excluindo valores faltantes da coluna Genero

```python
# Excluindo
df.dropna(subset=['Genero'], inplace=True)
#InplaceTrue faz com que as alterações sejam feitas diretamente no DataFrame original
```

```python
df['Genero'].isnull().sum()
```

a exclusão de dados faltantes na coluna Genero nao gera impacto na nossa base de dados já que a informação principal que queremos e Churn

```python
#Removendo coluna customerID
df.drop(columns=['customerID'], inplace=True)
```

Excluindo coluna customerID pois não são dados relevantes dentro do nosso modelo

CORRIGINDO DADOS INCONSISTENTES


```python
df['Servico_Internet'].unique()
```

```python
#Transformando todos os dados da coluna Servico_Internet em caixa alta
df['Servico_Internet'] = df['Servico_Internet'].str.upper()
```

```python
#Utilizando a função replace
df['Servico_Internet'] = df['Servico_Internet'].replace('DSL', 'DLS')
```

```python
#Visualizando alterações
df['Servico_Internet'].unique()
```

Os dados errados foram subistituido pela correção em letras maiusculas utilizando o comando upper

PADRONIZANDO COLUNAS TODAS EM PORTUGUÊS


```python
#Utilizando a função rename
df.rename(columns={'Dependents': 'Dependentes'}, inplace=True)
df.rename(columns={'PhoneService': 'Servico_Telefonico'}, inplace=True)
df.rename(columns={'StreamingTV': 'Transmissao_TV'}, inplace=True)
df.rename(columns={'PaymentMethod': 'Forma_Pagamento'}, inplace=True)
```
