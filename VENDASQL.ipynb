```python
import pandas as pd
import sqlite3

from google.colab import files
uploaded = files.upload()
```

```python
df_vendas = pd.read_csv("TB_VENDAS_TAREFA.csv", delimiter=';')
```

```python
conn = sqlite3.connect(':memory:')
df_vendas.to_sql('tb_vendas', conn, index=False , if_exists='replace')
```

```python
def run_query(query):
  return pd.read_sql(query, conn)
```

Faça uma consulta que retorne os nomes dos produtos distintos que temos na base de volta.

```python
query = """
SELECT
DISTINCT(PRODUTO)
FROM TB_VENDAS
"""
result_df = run_query(query)
print(result_df)
```

Faça uma consulta que retorne a contagem dos clientes distintos que temos na nossa base, não esqueça de renomear sua coluna.

```python
query = """
SELECT
COUNT(DISTINCT ID_CLIENTE) AS CLIENTE_DISTINTOS
FROM TB_VENDAS
"""
result_df = run_query(query)
print(result_df)
```

Faça um consulta que retorne uma coluna com os produtos distintos e o  valor_unid de cada produto. Porém apenas para produtos onde ovalor_unid é maior ou igual a 50 reis.

```python
query = """
SELECT
DISTINCT PRODUTO,
VALOR_UNID
FROM TB_VENDAS
WHERE VALOR_UNID >= 50
"""
result_df = run_query(query)
print(result_df)
```

Faça um consulta que retorne o ID das compras e o valor total gasto nessa compra (Valor_unidade X Unidades_) das 5 compras com mais valor total gasto.

```python
query = """
SELECT
VALOR_UNID * UNIDADES AS VALOR_TOTAL,
ID_COMPRA
FROM TB_VENDAS
ORDER BY VALOR_TOTAL DESC
LIMIT 5

"""
result_df = run_query(query)
print(result_df)
```

Faça uma consulta que retorne os produtos e a média do preço da unidade dos produtos, ordenando do maior para o menor.

```python
query = """
SELECT
PRODUTO,
AVG(VALOR_UNID) AS MEDIA_VALOR_UNID
FROM TB_VENDAS
GROUP BY PRODUTO
ORDER BY MEDIA_VALOR_UNID DESC
"""
result_df = run_query(query)
print(result_df)
```

Faça um consulta que retorne o id dos 3 clientes da base de vendas que maisrealizaram compras e a quantidade de compras realizadas.

```python
query = """
SELECT
ID_CLIENTE,
COUNT(ID_COMPRA) AS QUANTIDADE_COMPRAS
FROM TB_VENDAS
GROUP BY ID_CLIENTE
ORDER BY QUANTIDADE_COMPRAS DESC
LIMIT 3
"""
result_df = run_query(query)
print(result_df)
```
