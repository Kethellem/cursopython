```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm
```

Os dados são criados a seguir:

```python
#Definindo média e desvios padrão para as notas estratégias
media_estrategica_A = 70
desvio_padrao_estrategica_A = 10

media_estrategica_B =75
desvio_padrao_estrategica_B = 12

#Gerando as amostra de notas para cada estratégia de ensino da nossa base
np.random.seed(0) #Para garantir a reprodutibilidade dos resultados
amostra_estrategica_A = np.random.normal(loc=media_estrategica_A, scale=desvio_padrao_estrategica_A, size=50)
amostra_estrategica_B = np.random.normal(loc=media_estrategica_B, scale=desvio_padrao_estrategica_B, size=50)

print("Notas da Estratégia A:", amostra_estrategica_A[:5])
print("Notas da Estratégia B:", amostra_estrategica_B[:5])

```

1 - De acordo com as informaçoes analisada o nosso teste é unilateral á direita , esquerda ou bicauldal ? Justifique.

O teste é bicaudal, pois a hipótese alternativa considera que as médias são diferentes, sem especificar uma direção (maior ou menor).

2 - Calcule as médias para as duas amostragens e as variâncias. Quais insights você pode retirar comparando os dado ?

```python
#Realizando teste de hipótese
t_statistic, p_value = stats.ttest_ind(amostra_estrategica_A, amostra_estrategica_B)

#Imprimindo os resultados da variável criada
print("Estatística de teste (t-statistic):", t_statistic)
print("p-value:", p_value)
```

Estratégia A tenha uma média mais alta, essa diferença pode ter ocorrido por acaso, dado o tamanho pequeno da amostra (apenas 5 notas por grupo).


3 -Imprima os resultados da estatística do teste Z, p value e indique se rejeitaremos ou não a hipótese nula. Justifique sua resposta.

```python
nivel_significancia = 0.05

#Calculando as médias das amostras
if p_value < nivel_significancia:
    print("Rejeitamos a hipótese nula. As médias das amostras são diferentes.")
else:
    print("Não rejeitamos a hipótese nula. Não há evidências suficientes para afirmar que as médias das amostras são diferentes.")

```

 com base nessas amostras, não podemos afirmar que existe uma diferença significativa entre as médias das notas dos alunos das duas estratégias de ensino.



4 - Para finalizar monte um gráfico da distribuição de estatística do nosso teste z e explique o que pode ser observado através dele.  

```python
#Calculando média e desvio padrão das taxas de conversão
media_estrategica_A = np.mean(amostra_estrategica_A)
desvio_padrao_estrategica_A = np.std(amostra_estrategica_A)
media_estrategica_B = np.mean(amostra_estrategica_B)
desvio_padrao_estrategica_B = np.std(amostra_estrategica_B)

# Abaixo criamos um intervalo de valores para plotagem
x = np.linspace(40, 100, 100)  # intervalo de 40 a 100 com 100 pontos igualmente espaçados

# Plotando a densidade de probabilidade das taxas de conversão
plt.plot(x, norm.pdf(x, media_estrategica_A, desvio_padrao_estrategica_A), label='Estratégia A')
plt.plot(x, norm.pdf(x, media_estrategica_B, desvio_padrao_estrategica_B), label='Estratégia B')
plt.xlabel('Notas')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição das Notas')
plt.legend()
plt.show()
```
